```
public class Employee
{
    public string Name { get; set; }
    public string Age { get; set; }
    public string Salary { get; set; }
    public string WStatus { get; set; }
    public string VStatus { get; set; }
    public string Position { get; set; }
    public string Department { get; set; }

    public Employee(string name, string age, string salary, string wstatus, string vstatus, string position, string department)
    {
        Name = name;
        Age = age;
        Salary = salary;
        WStatus = wstatus;
        VStatus = vstatus;
        Position = position;
        Department = department;
    }

    public virtual string GetInfo()
    {
        return $"Имя: {Name}, Возраст: {Age}, ЗП: {Salary}, Должность: {Position}, Отдел: {Department}";
    }

    public virtual string Work()
    {
        return $"Статус задач: {WStatus}";
    }

    public virtual string TakeVacation()
    {
        return $"Статус отпуска: {VStatus}";
    }

    public virtual string GetDepartmentInfo()
    {
        return $"Отдел: {Department}";
    }
}

public class Manager : Employee
{
    public string SubordinatesCount { get; set; }
    public int MeetingsScheduled { get; set; }

    public Manager(string name, string age, string salary, string wstatus, string vstatus, 
                   string subordinatesCount, int meetingsScheduled) 
        : base(name, age, salary, wstatus, vstatus, "Менеджер", "Управление")
    {
        SubordinatesCount = subordinatesCount;
        MeetingsScheduled = meetingsScheduled;
    }

    public override string Work()
    {
        return $"Статус: {WStatus}, Количество подчиненных: {SubordinatesCount}, Запланированные встречи: {MeetingsScheduled}";
    }

    public string ConductMeeting()
    {
        return "Проведение встречи...";
    }
}

public class Developer : Employee
{
    public string Specialization { get; set; }
    public int ProjectsCompleted { get; set; }

    public Developer(string name, string age, string salary, string wstatus, 
                     string vstatus, string specialization, int projectsCompleted) 
        : base(name, age, salary, wstatus, vstatus, "Разработчик", "Разработка")
    {
        Specialization = specialization;
        ProjectsCompleted = projectsCompleted;
    }

    public override string TakeVacation()
    {
        return $"{Specialization} - Статус отпуска: {VStatus}";
    }

    public void CodeReview()
    {
        Console.WriteLine("Проведение ревью кода...");
    }
}

public class Designer : Employee
{
    public string DesignStyle { get; set; }
    public int ProjectsInProgress { get; set; }

    public Designer(string name, string age, string salary, 
                    string wstatus, string vstatus, string designstyle,
                    int projectsInProgress) 
        : base(name, age, salary, wstatus, vstatus, "Дизайнер", "Дизайн")
    {
        DesignStyle = designstyle;
        ProjectsInProgress = projectsInProgress;
    }

    public override string Work()
    {
        return $"Стиль: {DesignStyle}, Статус задач: {WStatus}, Проекты в работе: {ProjectsInProgress}";
    }

   public void CreateMockup()
   {
       Console.WriteLine("Создание макета...");
   }
}

public class TeamLead : Manager
{
   public TeamLead(string name, string age, string salary,
                   string wstatus, string vstatus,
                   string subordinatesCount,
                   int meetingsScheduled) 
       : base(name, age, salary, wstatus, vstatus,
              subordinatesCount, meetingsScheduled)
   {
   }

   public void MentorTeam()
   {
       Console.WriteLine("Наставничество команды...");
   }
}

public interface IReportable
{
   void GenerateReport();
}

public class ReportingManager : Manager, IReportable
{
   public ReportingManager(string name,
                           string age,
                           string salary,
                           string wstatus,
                           string vstatus,
                           string subordinatesCount,
                           int meetingsScheduled): base(name, age, salary, wstatus, vstatus,
              subordinatesCount, meetingsScheduled)
   {
   }

   public void GenerateReport()
   {
       Console.WriteLine("Генерация отчета...");
   }
}

Manager _manager = new Manager("Азимбек", "46", "16859 руб", "63%", "В процессе работы", "2", 5);
Console.WriteLine(_manager.GetInfo());
Console.WriteLine(_manager.TakeVacation());
Console.WriteLine(_manager.Work());
Console.WriteLine(_manager.ConductMeeting());
Console.WriteLine();

Developer _developer = new Developer("Всеволод", "76", "83650 руб", "100%", "В отпуске", "Графический разработчик", 10);
Console.WriteLine(_developer.GetInfo());
Console.WriteLine(_developer.TakeVacation());
_developer.CodeReview();
Console.WriteLine();

Designer _designer = new Designer("Джошуа", "27", "56235 руб", "100%", "В отпуске", "Барокко", 3);
Console.WriteLine(_designer.GetInfo());
Console.WriteLine(_designer.TakeVacation());
Console.WriteLine(_designer.Work());
_designer.CreateMockup();
Console.WriteLine();

TeamLead _teamLead = new TeamLead("Равшан", "35", "120000 руб", "75%", "В процессе работы", "5", 3);
Console.WriteLine(_teamLead.GetInfo());
Console.WriteLine(_teamLead.Work());
_teamLead.MentorTeam();
Console.WriteLine();

ReportingManager _reportingManager = new ReportingManager("Рубик", "40", "150000 руб", "80%", "В процессе работы", "8", 4);
Console.WriteLine(_reportingManager.GetInfo());
Console.WriteLine(_reportingManager.Work());
_reportingManager.GenerateReport();
```
